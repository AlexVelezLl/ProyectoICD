```python
import pandas as pd
import geopandas as gpd
```

```python
df_deliveries=pd.read_csv('../../data/data2/processed/steps.csv')
```

```python
df_deliveries.columns
```

```python
gdf_deliveries = gpd.GeoDataFrame(df_deliveries, geometry=gpd.points_from_xy(df_deliveries["lng"], df_deliveries["lat"]))
```

```python
gdfBH = gpd.read_file('../../data/BH_regions.geojson', driver='GeoJSON')
```

```python
dfgeo = gpd.sjoin(gdf_deliveries, gdfBH, how='inner', op='within')
```

```python
columns_to_take=['route_id', 'delivery_id', 'action', 'step_order',
       'distance_from_previous', 'total_size', 'cost', 'final_price',
       'shipper_id', 'effective_dt', 'lat', 'lng', 'shipper_capacity',
       'shipper_deliveries_completed', 'route_deliveries_count',
       'time_from_previous','CODIGO','NOME']
```

```python
dfgeo[columns_to_take].to_csv('../../data/data2/processed/steps_with_regions.csv',index=False)
```

```python
dfgeo[columns_to_take]
```

```python
gdfgeo = dfgeo.groupby('NOME')["delivery_id"].count().reset_index()
```

```python
gdfgeo = gdfBH.merge(gdfgeo, on='NOME')
```
